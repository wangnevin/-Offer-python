剑指 Offer 40. 最小的k个数

输入整数数组 arr ，找出其中最小的 k 个数。例如，输入4、5、1、6、2、7、3、8这8个数字，则最小的4个数字是1、2、3、4。


示例 1：
输入：arr = [3,2,1], k = 2
输出：[1,2] 或者 [2,1]


示例 2：
输入：arr = [0,1,2,1], k = 1
输出：[0]


限制：
	0 <= k <= arr.length <= 10000
	0 <= arr[i] <= 10000

可以基于Partition函数来解决这个问题。

基于数组的第k个数字来调整，使得:
* 1.比第k个数字小的所有数字都位于数组的左边
* 2.比第k个数字大的所有数字都位于数组的右边

这样调整之后，位于数组中左边的k个数字就是最小的k个数字（这k个数字不一定是排序的）。

不足：采用这种思路是有限制的。我们需要修改输入的数组，因为函数Partition会调整数组中数字的顺序。

```python
import random

class Solution:
    def getLeastNumbers(self, arr: List[int], k: int) -> List[int]:
        if arr == None or k > len(arr) or k <= 0:
            return []

        n = len(arr)
        start = 0
        end = n - 1
        
        index = self.Partition(arr, n, start, end)

        while index != k - 1:
            if index > k - 1:
                end = index - 1
                index = self.Partition(arr, n, start, end)
            else:
                start = index + 1
                index = self.Partition(arr, n, start, end)
        
        result = []
        for i in range(k):
            result.append(arr[i])
        
        return result
    
    def Partition(self, arr, n, start, end):
        if arr == None or n <= 0 or start < 0 or end >= n:
            return
        
        # 随机选中一个数，然后将它与最后一个数交换位置
        index = random.randint(start, end)
        self.swap(arr, index, end)

        small = start - 1

        # 遍历数组（不包括随即选中的数）
        # 把每一个小于随机选中的数的数放到数组靠前的部分
        for i in range(start, end):
            if arr[i] < arr[end]:
                small += 1
                if small != i:
                    self.swap(arr, i, small)
        small += 1
        
        # 把选中的数与small指向的数交换，此时选中的数就是第small小的数
        self.swap(arr, small, end)
        
        return small

    def swap(self, arr, i, j):
        tmp = arr[i]
        arr[i] = arr[j]
        arr[j] = tmp
        return arr
```

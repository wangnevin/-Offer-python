剑指 Offer 53 - I. 在排序数组中查找数字 I

统计一个数字在排序数组中出现的次数。


示例 1:
输入: nums = [5,7,7,8,8,10], target = 8
输出: 2

示例 2:
输入: nums = [5,7,7,8,8,10], target = 6
输出: 0


限制：0 <= 数组长度 <= 50000

我们先分析如何用二分查找算法在数组中找到第一个k。

二分查找算法总是先拿数组中间的数字和k作比较：
* 1.如果中间的数字比k大，那么k只有可能出现在数组的前半段，下一轮我们只在数组的前半段查找就可以了。
* 2.如果中间的数字比k小，那么k只有可能出现在数组的后半段，下一轮我们只在数组的后半段查找就可以了。
* 3.如果中间的数字和k相等，我们先判断这个数字是不是第一个k：
    * a.如果中间数字的前面一个数字不是k，此时中间的数字刚好就是第一个k。
    * b.如果中间数字的前面一个数字也是k，也就是说第一个k肯定在数组的前半段，下一轮我们仍然需要在数组的前半段查找。

之后用同样的方法在数组中找到最后一个k：
* 1.如果中间数字比k大，那么k只能出现在数组的前半段。
* 2.如果中间数字比k小，k就只能出现在数组的后半段。
* 3.如果中间数字等于k，我们需要判断这个k是不是最后一个k，也就是中间数字的下一个数字是不是也等于k：
    * a.如果下一个数字不是k，则中间数字就是最后一个k了；
    * b.否则下一轮我们还是要在数组的后半段中去查找。

在分别找到第一个k和最后一个k的下标之后，我们就能计算出k在数组中出现的次数了。

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        count = 0
        first = -1
        last = -1
        if nums != None and len(nums) > 0:
            first = self.getFirstK(nums, 0, len(nums)-1, target)
            last = self.getLastK(nums, 0, len(nums)-1, target)
        
        if first > -1 and last > -1:
            count = last - first + 1
        
        return count
    
    def getFirstK(self, nums, start, end, target):
        if start > end:
            return -1

        middle = int((end + start)/2)
        # 如果中间的数字和k相等，我们先判断这个数字是不是第一个k
        if nums[middle] == target:
            if middle == 0 or (middle > 0 and nums[middle-1] != target):
                return middle
            else:
                end = middle - 1
        
        # 如果中间的数字比k小，那么k只有可能出现在数组的后半段
        elif nums[middle] < target:
            start = middle + 1
        
        # 如果中间的数字比k大，那么k只有可能出现在数组的前半段
        else:
            end = middle - 1
        
        return self.getFirstK(nums, start, end, target)
    
    def getLastK(self, nums, start, end, target):
        if start > end:
            return -1
        
        middle = int((end+start)/2)
        if nums[middle] == target:
            if middle == len(nums)-1 or (middle<len(nums)-1 and nums[middle+1] != target):
                return middle
            else:
                start = middle + 1
        elif nums[middle] < target:
            start = middle + 1
        else:
            end = middle - 1
        
        return self.getLastK(nums, start, end, target)
```

# 剑指 Offer 18. 删除链表的节点

给定单向链表的头指针和一个要删除的节点的值，定义一个函数删除该节点。

返回删除后的链表的头节点。

注意：此题对比原题有改动

示例 1:
输入: head = [4,5,1,9], val = 5
输出: [4,1,9]
解释: 给定你链表中值为 5 的第二个节点，那么在调用了你的函数之后，该链表应变为 4 -> 1 -> 9.

示例 2:
输入: head = [4,5,1,9], val = 1
输出: [4,5,9]
解释: 给定你链表中值为 1 的第三个节点，那么在调用了你的函数之后，该链表应变为 4 -> 5 -> 9.


说明：
	题目保证链表中节点的值互不相同
	若使用 C 或 C++ 语言，你不需要 free 或 delete 被删除的节点

方法：把要删除的节点的下一个节点的val和next都赋值给要删除的节点，再删除要删除节点的下一个节点

依照以上方法，要分4种情况：
* 1.只有头节点，并且删除头节点
* 2.有多个节点，删除头节点
* 3.有多个节点，删除尾节点
* 4.有多个节点，删除中间的节点（不是尾节点也不是头节点）

对应的解决方案：
* 1.直接令head=None即可
* 2.把头节点的下一个节点的val和next都赋值给头节点，再删除头节点的下一个节点
* 3.从头开始遍历链表，找到尾节点的前一个节点，令其next=None
* 4.从头开始遍历链表，找到要删除的节点，把要删除的节点的下一个节点的val和next都赋值给要删除的节点，再删除要删除节点的下一个节点

注意：
第3和第4种情况都需要遍历列表。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None
class Solution:
    def deleteNode(self, head: ListNode, val: int) -> ListNode:
        # 如果只有一个节点（头节点），删除头节点
        if head.val == val and head.next == None:
            head = None
            return head
        
        # 有多个节点，删除头节点
        if head.val == val:
            next_node = head.next
            head.val = next_node.val
            head.next = next_node.next
            next_node = None
            return head
        
        # 有多个节点，删除的不是头节点（要删除的节点有上一个节点）
        NodeToBeDeleted = None
        current = head
        find_target = False
        while True:
            if current.next.val == val:
                pre_node = current
                NodeToBeDeleted = current.next
                find_target = True
                break
            current = current.next
            if current.next == None:
                break
        
        if find_target:
            # 要删除的是尾节点
            if NodeToBeDeleted.next == None:
                pre_node.next = None
                return head
            
            # 要删除的不是尾节点
            elif NodeToBeDeleted.next != None:
                next_node = NodeToBeDeleted.next
                NodeToBeDeleted.val = next_node.val
                NodeToBeDeleted.next = next_node.next
                next_node = None
                return head
```

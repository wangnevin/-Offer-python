# 剑指 Offer 07. 重建二叉树

输入某二叉树的前序遍历和中序遍历的结果，请重建该二叉树。假设输入的前序遍历和中序遍历的结果中都不含重复的数字。

例如，给出

前序遍历 preorder = [3,9,20,15,7]
中序遍历 inorder = [9,3,15,20,7]
返回如下的二叉树：

    3
   / \
  9  20
    /  \
   15   7

对于前序遍历，第一个元素即根节点的值。
对于中序遍历，根节点的值的左边的值是根节点的左子树的值，右边的值是右子树的值。

比如上例，根节点的值是3，3是中序遍历的第二个值，3的左边是[9]，右边是[15,20,7]，那么左子树中的节点的值为9，右子树中的节点的值为15，20和7。在前序遍历列表中，根节点3后面首先跟的是左子树的节点的值[9]，然后跟的是右子树节点的值[20,15,7]。

有了以上规律后，我们就有如下算法：
*    1.为前序遍历列表和中序遍历列表都设两个指针，分别指向头和尾，即startPreorder, endPreorder, startInorder, endInorder。
*    2.首先根据前序遍历的第一个元素找打根节点的值rootVal。
*    3.然后在中序遍历列表中找到rootVal的位置rootInorder。
*    4.计算出左子树中的节点数量，即leftLength=rootInorder-startInorder。
*    5.构建左子树时，令：
*        (1).startPreorder指向当前值的下一个值startPreorder+1；
*        (2).endPreorder指向前序遍历中当前左子树的最后一个节点startPreorder+leftLength；
*        (3).startInorder不变；
*        (4).endInorder指向当前根节点的左边一个节点rootInorder-1
*    6.构建右子树时，令：
*        (1).startPreorder指向前序遍历中当前右子树的第一个节点startPreorder+leftLength；
*        (2).endPreorder不变
*        (3).startInorder指向当前根节点的右边一个节点rootInorder+1
*        (4).endInorder不变

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        if preorder == None or len(preorder) == 0:
            return None
        if inorder == None or len(inorder) == 0:
            return None
        
        startPreOrder = 0
        endPreOrder = len(preorder) - 1
        startInOrder = 0
        endInOrder = len(inorder) - 1

        return self.helper(preorder, inorder, startPreOrder, endPreOrder, startInOrder, endInOrder)

    def helper(self, preorder, inorder, startPreOrder, endPreOrder, startInOrder, endInOrder):
        # 当前前序遍历的首元素即当前二叉树的根节点的值
        rootVal = preorder[startPreOrder]
        root = TreeNode(rootVal)
        
        # 在中序遍历数组中找出当前根节点的值的位置(rootInorder)
        for i in range(len(inorder)):
            if inorder[i] == rootVal:
                rootInOrder = i     
        
        # 中序遍历遍历列表的根节点的值的左边的数是当前根节点的左子树的值，右边的数是右子树的值
        # 先用rootInorder-startInorder找出左子树的节点数量(leftLength)
        # 再在前序遍历列表中找到当前根节点左子树的起始位置
        leftLength = rootInOrder - startInOrder
        rightLength = endInOrder - rootInOrder
        endLeftPreOrder = startPreOrder + leftLength
        startRightPreOrder = endLeftPreOrder + 1

        # 构建左子树
        if leftLength > 0:
            root.left = self.helper(preorder, inorder, startPreOrder+1, endLeftPreOrder, startInOrder, rootInOrder-1)
        # 构建右子树
        if rightLength > 0:
            root.right = self.helper(preorder, inorder, startRightPreOrder, endPreOrder, rootInOrder+1, endInOrder)
        
        return root
```
